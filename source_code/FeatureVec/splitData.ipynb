```python
"""
 author:jjk
 datetime:2019/12/01
 coding:utf-8
 project name:Pycharm_workstation
 Program function:   解析数据文件，返回特征集和标签
 
"""

'''
特征数据集：
特征1：年均投入时间（min）
特征2：玩游戏占时间百分比
特征3：每天看综艺的时间（h）
标签集：
1：学习专注
2：学习正常
3：比较贪玩
'''
import os
from numpy import *
```

```python
def file_matrix(filepath):
    file = open(filepath,'r',encoding='utf-8')
    arrayLines = file.readlines()
    # print(arrayLines)
    returnMat = zeros((len(arrayLines),3)) # len(arraylines)行，3列——特征数据集
    # print(returnMat)
    
    classLabelMat = [] # 标签集
    index = 0 # 索引
    for line in arrayLines: 
        listForLine = line.strip().split('\t')
        returnMat[index,:] = listForLine[0:3] # 获取前散列
        classLabelMat.append(int(listForLine[-1])) # 添加最后一个字段
        index += 1
    file.close()
    return returnMat,classLabelMat # 返回：特征矩阵，标签矩阵
    
```

```python
if __name__ == '__main__':
    filepath = os.path.abspath(r'../dataSet/files/dataset.txt')
    returnMat,classLabelMat = file_matrix(filepath) # 函数实例化
    print('数据集:\n',returnMat,'\n标签集：\n',classLabelMat)
    
    
    
```
