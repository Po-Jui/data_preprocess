```python
# 利用sklearn计算tfidf值特征
def sklearn_tfidf_feature(corpus=None):
    # 该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频
    vectorizer = CountVectorizer() # 构建词汇表
    transformer = TfidfTransformer() # 该类会统计每个词语tf-idf权值
    
    tfidf = transformer.fit_transform(vectorizer.fit_transform(corpus)) # 处理后的词向量填充进来，然后统计它的tf-idf权值
    # print(tfidf)
    
    words = vectorizer.get_feature_names() # 获取词袋模型中的所有的词语
    # 将tf-idf矩阵抽取出来，元素a[i][j]表示j词在i类文本中的tf-idf权重
    weight = tfidf.toarray()
    #print(words) # 两篇文章集中到一起，而且是没有重复的
    #print(weight) # 权重
    
    # 打印每类文本的tf-idf词语权重，第一个for遍历所有文本，第二个for便利某一类文本下的词语权重
    for i in range(len(weight)):
        print(u"-----这里输出第",i,u"类文本的词语tf-idf权重")
        for j in range(len(words)):
            print(words[j],weight[i][j])
        
```

```python
if __name__ == '__main__':
    corpus = []
    path1 = r'../dataSet/CSCMNews/体育/2.txt'
    str_doc1 = readFile(path1) # 实例化：读取文本内容
    word_list1 = ' '.join(seg_doc(str_doc1))
    #print(word_list1)
    
    path2 = r'../dataSet/CSCMNews/时政/339764.txt'
    str_doc2 = readFile(path2) # 实例化：读取文本内容
    word_list2 = ' '.join(seg_doc(str_doc2))
    #print(word_list2)
    
    # 以上是经过分词处理之后的两类词矩阵/词向量
    corpus.append(word_list1)
    corpus.append(word_list2) # 两类都装在一个列表下
    #print(corpus)
    sklearn_tfidf_feature(corpus) # 调用sklearn计算tfidf值特征函数；第0类：体育新闻；第1类：时政新闻；每个词出现的频率
    
    
    
    
```
